```python
import xml.etree.ElementTree as ET
import re
import spacy
from spacy import displacy
from spacy.matcher import Matcher
from spacy.matcher import PhraseMatcher
```

```python
tree = ET.parse('../Data/19021.xml')
root = tree.getroot()
for child in root:
    print(child.tag, child.attrib)
```

```python
for content in root.iter('TEXT'):
    meeting_content = content.text


oneline = " ".join(line.strip() for line in meeting_content.splitlines())
```

```python
nlp = spacy.load("de_core_news_sm")
doc = nlp(oneline)
matcher = Matcher(nlp.vocab)
```

```python
for token in doc.ents:
    print(token.text, token.label_)
```

```python
pattern = [ {"ENT_TYPE": "PER"}, {"TEXT": '(', }, {"ENT_TYPE": "ORG"}, {"TEXT": ')' }]
matcher.add("TestAntrag", [pattern])
```

```python
matches = matcher(doc)
```

```python
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(match_id, string_id, start, end, span.text)
```
