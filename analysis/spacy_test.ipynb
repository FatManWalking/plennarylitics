```python
for content in root.iter('TEXT'):
    meeting_content = content.text
```

```python
split = meeting_content.split("Liste der entschuldigten Abgeordneten")
split = split[2].split("Anlage")
missing_mps = split[0].split("\n")
print(missing_mps)
```

```python
missing_mps_DIE_LINKE = []
missing_mps_CDUCSU = []
missing_mps_FDP = []
missing_mps_SPD = []
missing_mps_GRUENE = []
missing_mps_FRAKTIONSLOS = []
missing_mps_AFD = []

for element in missing_mps:
    if "DIE LINKE" in element:
        element = element.split("DIE LINKE")
        missing_mps_DIE_LINKE.append(element[0])
        
    if "CDU/CSU" in element:
        element = element.split("CDU/CSU")
        missing_mps_CDUCSU.append(element[0])

    if "FDP" in element:
        element = element.split("FDP")
        missing_mps_FDP.append(element[0])

    if "SPD" in element:
        element = element.split("SPD")
        missing_mps_SPD.append(element[0])
    
    if "BÜNDNIS 90/DIE GRÜNEN" in element:
        element = element.split("BÜNDNIS 90/DIE GRÜNEN")
        missing_mps_GRUENE.append(element[0])
    
    if "fraktionslos" in element:
        element = element.split("fraktionslos")
        missing_mps_FRAKTIONSLOS.append(element[0])

    if "AfD" in element:
        element = element.split("AfD")
        missing_mps_AFD.append(element[0])
        

print(missing_mps_DIE_LINKE)
print(missing_mps_CDUCSU)
print(missing_mps_FDP)
print(missing_mps_SPD)
print(missing_mps_GRUENE)
print(missing_mps_FRAKTIONSLOS)
print(missing_mps_AFD)
   
    
```
