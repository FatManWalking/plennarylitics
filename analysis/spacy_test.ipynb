```python
element = " ".join(line.strip() for line in meeting_content.splitlines())
element = element.replace(":", "")
element = element.replace(".", "")


```

```python
nlp = spacy.load("de_core_news_sm")
doc = nlp(element)
matcher = Matcher(nlp.vocab)
```

```python
pattern = [{"LOWER": "antrag"}, {"LOWER": "der"}, {"LOWER": "abgeordneten"}]
matcher.add("TestAntrag", [pattern])
```

```python
matches = matcher(doc)
```

```python
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(match_id, string_id, start, end, span.text)
```

```python


phrase_matcher = PhraseMatcher(nlp.vocab)
phrases = ['Antrag der Abgeordneten']
patterns = [nlp(text) for text in phrases]
phrase_matcher.add('Antrag', None, *patterns)

for sent in doc.sents:
    for match_id, start, end in phrase_matcher(nlp(sent.text)):
        if nlp.vocab.strings[match_id] in ["Antrag"]:
            print(sent.text)
```
